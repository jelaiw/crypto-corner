```python
import os
from cryptography.hazmat.primitives.ciphers.aead import AESGCM
```

```python
plaintext = b'a secret message'
```

```python
aad = b'authenticated but unencrypted data'
```

```python
key = AESGCM.generate_key(bit_length=128)
```

```python
aesgcm = AESGCM(key)
```

```python
nonce = os.urandom(12)
```

```python
c = aesgcm.encrypt(nonce, plaintext, aad)
```

```python
from binascii import hexlify as hexa
print(hexa(c))
```

```python
aesgcm.decrypt(nonce, c, aad)
```

```python
# For further detail, see https://cryptography.io/en/latest/hazmat/primitives/aead/#cryptography.hazmat.primitives.ciphers.aead.AESGCM.
# Force failure to verify MAC.
aesgcm.decrypt(nonce, c, b'foo')
```
