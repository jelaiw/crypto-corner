```python
from Crypto.Hash import Poly1305
from Crypto.Cipher import AES
```

```python
secret = b'Thirtytwo very very secret bytes'
```

```python
mac = Poly1305.new(key=secret, cipher=AES)
```

```python
mac.update(b'Hello')
```

```python
print("nonce: ", mac.nonce.hex())
```

```python
print("MAC tag: ", mac.hexdigest())
```
